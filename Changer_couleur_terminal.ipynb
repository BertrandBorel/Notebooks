# <center> Changer les couleurs dans un terminal </center>

Ouvrir un terminal et entrez la commande suivante : `COLOR [attribut]` où [attribut] signifie la couleur que vous voulez.

Les différents attributs sont : 
- 0 = Noir        
- 1 = Bleu        
- 2 = Vert        
- 3 = Bleu-gris   
- 4 = Rouge      
- 5 = Violet    
- 6 = Jaune       
- 7 = Blanc      
- 8 = Gris
- 9 = Bleu clair

Et : 
- A = Vert clair
- B = Cyan
- C = Rouge clair
- D = Violet clair
- E = Jaune clair
- F = Blanc brillant

Par exemple `COLOR 6` changera la couleur de la police en jaune.
