## <center> Ajouter des colonnes </center>

On va ajouter une colonne 'price_total' pour cela il faut nettoyer les données des valeurs nulles/manquantes et remplacer les $ pour convertir la colonne 'price_paid' en float.

```python
df.dropna(subset=["tax"], inplace=True)
```

```python
df
```

## Remplacer les $

```python
df['price_paid'] = df['price_paid'].str.replace('$', "")
```

```python
df['price_paid']
```

## Transformer le type d'une colonne

```python
# on transforme la colonne price_paid en float

# on remplace le $
df.price_paid = df.price_paid.apply(lambda x: x.replace("$", ""))

df.price_paid = df.price_paid.astype(float)
df.dtypes
```

```python
# on crée une colonne "price_total" 
df["price_total"] = df["price_paid"] * (1 - df["tax"] / 100)
```

```python
df
```

```python
# On remplace les pays par des abréviations avec la méthode map (c'est du mapping). Ex : France devient FR
countries = {"United States": "UN", "France": "FR", "Canada": "CA", "Morocco": "MA"}
df["country_code"] = df["country"].map(countries)
```

```python
df
```
