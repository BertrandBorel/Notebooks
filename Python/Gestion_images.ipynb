## Ouvrir une image avec pillow

Ouvre l'image dans l'application dédiée.

```python
from PIL import Image
  
# open method used to open different extension image file
im = Image.open('picture.png')
  
# This method will show image in any image viewer 
im.show() 
```

## Ouvrir une image avec matplotlib

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
 
# Read Images
img = mpimg.imread('picture.png')
 
# Output Images
plt.imshow(img)
```

## Création de thumbnails (vignettes)

```python
from PIL import Image
import glob, os

size = 128, 128

for infile in glob.glob("*.png"):
    file, ext = os.path.splitext(infile)
    with Image.open(infile) as im:
        im.thumbnail(size)
        im.save(file + "thumbnail.png")
```

## Parcourir les images d'un dossier

```python
# import required module
import os
# assign directory
directory = '../projet_df/train_data/train/'

# Sauvegarder les images dans une liste :
liste_images = []
 
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        print(f)
        liste_images.append(f)
```

## Afficher les images d'une liste

```python
# Read Images
for x in liste_images : 
    img = mpimg.imread(x)
 
    # Output Images
    plt.imshow(img)
    # affiche la première
    break
```

## Rotation d'image

```python
from PIL import Image
  
def main():
    try:
        #Relative Path
        img = Image.open("picture.png") 
          
        #Angle given
        img = img.rotate(180) 
          
         #Saved in the same relative location
        img.save("rotated_picture.png")
    except IOError:
        pass
  
if __name__ == "__main__":
    main()
```

```python
# Affichage de l'image modifiée :

# Read Images
img = mpimg.imread('rotated_picture.png')
 
# Output Images
plt.imshow(img)
```
