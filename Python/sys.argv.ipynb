# <center> Python : utilisation de `sys.argv` </center> 

En Python, sys.argv est une liste qui contient les arguments passés à un script Python à partir de la ligne de commande. Le premier élément de cette liste, `sys.argv[0]`, est le nom du script Python lui-même. Les éléments suivants, `sys.argv[1]`, `sys.argv[2]`, ... sont les arguments passés au script.

Voici un exemple d'utilisation de `sys.argv` pour passer des arguments à un script Python :

```python
import sys

# afficher le nom du script
print("Nom du script :", sys.argv[0])

# afficher les arguments passés au script
for arg in sys.argv[1:]:
    print("Argument :", arg)
```
